````python
from numpy import *
from scipy.interpolate import lagrange
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt

plt.style.use ('seaborn-poster')
````

# Método de LaGrange

## Implementación

``` lagrange_interpol(xi, yi, x0): ``` Implementación del método de LaGrange para interpolación polinómica

### Parámetros
- ``` xi ``` : conjunto de puntos *xi*
- ``` yi ``` : conjunto de puntos *yi*
- ``` x0 ``` : aproximación de la función en un punto determinado

````python
def lagrange_interpol(xi, yi, x0):
    result = 0

    for i in range(len(xi) ):
        l = 1

        for j in range(len(xi) ):
            if j != i:
                l = l * ( (x0 - xi[j] ) / (xi[i] - xi[j] ) )

        result = result + l * yi[i]

    return result
````

## Inserción de datos

````python
xi = [1, 1.05, 1.10, 1.15]
yi = [2.718282, 2.857651, 3.004166, 3.158193]
x0 = 1.07
````

## Salida de datos

````python
valor_interp = lagrange_interpol(xi, yi, x0) 
int_pol1d = lagrange(xi, yi)
pol_interp = poly.Polynomial(int_pol1d.coef[::-1] )

x = linspace(xi[0], xi[-1] )
fig = plt.figure(figsize = (10,8) )

plt.plot(x, pol_interp(x), 'b', xi, yi, 'ro', x0, pol_interp(x0), "go")
plt.title('Polinomio de LaGrange obtenido')
plt.xlabel('x')
plt.ylabel('y')
plt.grid ()
plt.show ()

print(f"Polinomio de LaGrange: {pol_interp}")
print(f"Valor obtenido al evaluar {x0}: {valor_interp}")
````
